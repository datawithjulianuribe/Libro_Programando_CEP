##Algoritmo de la carta de control P

```python
#Importe de la librería Numpy para el manejo de datos.
#Importe de la librería Matplotlib para la construcción del gráfico de control.
import numpy as np
import matplotlib.pyplot as plt
#Se deben importar la librería Statistics para el manejo de estadísticas básicas (otra librería adicional para explorar el manejo de datos)
#Se importa la librería Math
from statistics import mean, stdev
import math

#Calcular la proporción de piezas defectuosas en cada muestra. Para eso se utiliza una lista vacia y mediante un ciclo for se calculan las proporciones requeridas.
proporciones = []
n = 50 #Definir el tamaño de cada muestra
for i in defectos:
  proporciones.append(defectos[i]/n)

#Calcular la media de las proporciones y los límites de control, y aplicar un sí condicional para su control.
p_barra = mean(proporciones)
LSC = p_barra + 3 * math.sqrt(p_barra*(1-p_barra)/n)
if (p_barra - 3 * math.sqrt(p_barra*(1-p_barra)/n)) < 0:
  LIC = 0
else:
  LIC = p_barra - 3 * math.sqrt(p_barra*(1-p_barra)/n)

#Presentar los resultados de los lîmites de control.
print("Media de proporciones:", p_barra)
print("Límite de control superior:", LSC)
print("Límite de control inferior:", LIC)

#Crear la carta de control P para los datos con los límites calculados con la librería Matplotlib.
fig, ax = plt.subplots()
ax.plot(proporciones, 'o-', label='Proporción de defectos')
ax.axhline(y=p_barra, color='r', linestyle='--', label='LC')
ax.axhline(y=LSC, color='g', linestyle='--', label='LSC')
ax.axhline(y=LIC, color='g', linestyle='--', label='LIC')
ax.set(title='Carta de control P', xlabel='Muestras', ylabel='Proporciones')
ax.legend()
plt.show()
```
