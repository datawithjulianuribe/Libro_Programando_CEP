#Ejercicio demostrativo Carta de control U

##Datos de prueba

```python
# Importe de la librería Numpy para el manejo de datos.
import numpy as np

# Datos de entrada.
# Se define la variable defectos como un arreglo.
# Número de defectos por unidad.
defectos=np.array([1, 0, 2, 3, 1, 2, 0, 1, 4, 2, 2, 1, 3, 0, 0, 1, 0, 1, 2, 1])
elementos=np.array([5, 8, 7, 5, 9, 6, 7, 9, 4, 9, 6, 4, 9, 8, 5, 6, 4, 9, 4, 9])
```

##Algoritmo de la carta de control U

```python
#Importe de la librería Numpy para el manejo de datos.
#Importe de la librería Matplotlib para la construcción del gráfico de control.
import numpy as np
import matplotlib.pyplot as plt
#Calcular el número promedio de defectos total.
defectos_prom_elementos = sum(defectos)/sum(elementos)
#Calcular el número promedio de defectos por elementos.
defectos_promedio = []
for i, j in zip(defectos, elementos):
  defectos_promedio.append(i/j)
#Calcular límites de control de la carta U.
LIC = []
for j in elementos:
  if (defectos_prom_elementos-3*np.sqrt(defectos_prom_elementos/j))<0:
    LIC.append(0)
  else:
    LIC.append(defectos_prom_elementos-3*np.sqrt(defectos_prom_elementos/j))
LSC = []
for j in elementos:
  LSC.append(defectos_prom_elementos+3*np.sqrt(defectos_prom_elementos/j))
LSC_ = [round(valor, 3) for valor in LSC]
#Presentar los resultados de los lîmites de control.
print("Media de defectos por lote:", round(defectos_prom_elementos, 3))
print("Límite de control superior:", LSC_)
print("Límite de control inferior:", LIC)
#Crear la carta de control np para los datos con los límites calculados
#con la librería Matplotlib.
fig, ax = plt.subplots()
ax.plot(defectos_promedio, 'o-', markersize=5, color='black', label='Defectos')
ax.plot(LSC_, color='r', linestyle='--', label='LSC')
ax.axhline(y=defectos_prom_elementos, color='k', linestyle='-', label='LC')
ax.plot(LIC, color='r', linestyle='--', label='LIC')
ax.set_title('Carta de control U para atributos')
ax.set_xlabel('Lotes')
ax.set_ylabel('Número de defectos por lote inspeccionado')
ax.legend(loc='upper right')
plt.show()
```
